## Pie Chart for Merged Data

```python
sport_counts = Merged_data['Sport'].value_counts()


def absolute_value(val):
    return int(val/100.*sum(sport_counts))

plt.pie(sport_counts,autopct=lambda p: f'{absolute_value(p)}', shadow=True, startangle=140)

plt.axis('equal')
plt.title('Pie chart for Number of Participants in Each Sport')
```

```python
Data = Merged_data
```

```python
df = pd.DataFrame(Data)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Medal', y='Age', data=df, palette='Set3')
plt.title('Box Plot of Age by Medal Type')
plt.xlabel('Medal')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
df_avg_age = df.groupby('Year')['Age'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_avg_age['Year'], df_avg_age['Age'], marker='o', linestyle='-', color='b')
plt.title('Average Age of Athletes by Year')
plt.xlabel('Year')
plt.ylabel('Average Age')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Sex', multiple='stack', kde=False, palette='Set2', bins=25)
plt.title('Histogram of Age by Sex')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

# Detailed Analysis of the Data


The Olympic dataset contains valuable information about athletes who competed in the Olympics over the years. In this analysis, we'll look at trends, patterns, and interesting insights from the data, supported by visualizations.

Trends in the Data
Medal Distribution Over Time:
We can see how the number of medals given out has changed over the years. In the early Olympics, fewer medals were awarded because there were fewer sports and countries involved. As the Olympics grew, more sports were added, and more countries participated, leading to more medals being won.
Some years might show fewer medals due to events like World Wars or political issues, which affected the Games.

Changes in Athlete Profiles:
It's also interesting to see how the average age, height, and weight of athletes have changed over time. As sports science improved, athletes might have become taller, heavier, or older because of better training and nutrition.
Different sports may have different trends. For example, endurance sports might have older athletes, while sports that require agility might have younger participants.

Gender Representation:
The data shows how the participation of men and women in the Olympics has changed. In the past, men were more dominant, but over time, more women started participating, especially as more women's events were introduced.
We can also see which sports have seen the most growth in female participation and which ones remain mostly male.

Patterns in the Data
Correlation Between Physical Attributes:
We can explore how height and weight are related, especially in different sports. For example, in basketball, athletes might be taller and heavier, while gymnasts might be shorter and lighter.
Understanding these relationships can help us see the physical demands of each sport. Sports that require strength might show a strong connection between height and weight, while speed-focused sports might not.

Age Distribution in Different Sports:
The age of athletes varies widely depending on the sport. For example, gymnasts are often younger because of the agility required, while equestrian events might have older participants.
This analysis can also highlight exceptional cases, like very young or very old athletes who have succeeded in their sport.

Identifying Outliers:
Outliers are unusual data points that stand out. In this dataset, outliers could be very young or old athletes or those with extreme physical attributes. Identifying these can help us understand unique stories or spot possible errors in the data.
For example, a very young athlete who won a medal could be a remarkable case in Olympic history.

Common Sense Insights
Global Participation:
The NOC (National Olympic Committee) data shows how global participation in the Olympics has changed. By looking at which countries participated and when, we can see how the Olympics have become more inclusive over time.
In the past, participation was mostly from Europe and North America, but now more countries from Asia, Africa, and South America are involved. This reflects how the Olympics have become a truly global event.

Sports Dominance by Country:
Certain countries excel in specific sports, and this dataset helps us identify these patterns. For example, the United States might dominate in swimming and athletics, while China might be strong in gymnastics.
Understanding these patterns can give us insights into why some countries are better at certain sports, which could be due to cultural, economic, or environmental factors.

Visualizations
Medal Trends Over Time:
A line plot can show the total number of medals given out each Olympic year. This helps us see trends over time, like the growth of the Olympics or years with fewer medals due to disruptions.

Pair Plot for Physical Attributes:
A pair plot (scatterplot matrix) can help us visualize the relationships between age, height, and weight in different sports. This plot gives us a clear view of how these physical characteristics vary across sports and can highlight any unusual patterns.

Conclusion:

This analysis of the Olympic dataset reveals some key trends and patterns. The Olympics have grown significantly over time, with more sports, events, and participants, leading to more medals being awarded. The profiles of athletes have also changed, reflecting improvements in training, nutrition, and the global spread of sports.Gender representation has improved, making the Olympics more inclusive. The physical attributes of athletes, such as age, height, and weight, vary depending on the sport, showing the different demands of each event.
